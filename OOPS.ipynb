```python
#1.

In object-oriented programming, a class is a blueprint that defines the structure and behavior of objects. 
An object is an instance of a class and represents a specific entity. 
For example, a class called "Car" would define attributes like make, model, and color, and methods like start, stop, accelerate, and brake. 
Objects created from the Car class would have their own attribute values and can call the defined methods.

#Example

class Car:
    def __init__(self, model, top_speed, color):
        self.model = model
        self.top_speed = top_speed
        self.color = color
        

    def car_details(self):
        return self.model, self.top_speed, self.color

# Creating objects of the Car class
my_car1 = Car("Toyota", 180, "Blue")
my_car2 = Car("Honda", 160, "Red")
my_car3 = Car("Suzuki", 190, "Black")

So, here, 'Car' is my class and my_car1, my_car2, and my_car3 are objects of the class 'Car'.
```

```python
#5.

Inheritance in object-oriented programming allows a derived class to inherit attributes and methods from a base class. 
There are different types of inheritance:

1. Single Inheritance: In single inheritance, a derived class inherits from a single base class. 
It forms a simple parent-child relationship between classes.

#Example

class Vehicle:
    def move(self):
        print("Vehicle is moving.")

class Car(Vehicle):
    def honk(self):
        print("Car is honking.")

my_car = Car()
my_car.move()  # Output: Vehicle is moving.
my_car.honk()  # Output: Car is honking.

2. Multiple Inheritance: Multiple inheritance allows a derived class to inherit from multiple base classes. 
It provides the ability to combine characteristics and behaviors from different classes into a single derived class.

#Example

class Animal:
    def eat(self):
        print("Animal is eating.")

class Flyable:
    def fly(self):
        print("Flying...")

class Bird(Animal, Flyable):
    def chirp(self):
        print("Bird is chirping.")

my_bird = Bird()
my_bird.eat()  # Output: Animal is eating.
my_bird.fly()  # Output: Flying...
my_bird.chirp()  # Output: Bird is chirping.

3. Multilevel Inheritance: Multilevel inheritance involves creating a chain of inheritance, where a derived class becomes the base class for another class. 
It forms a hierarchical structure of classes.

#Example

class Animal:
    def eat(self):
        print("Animal is eating.")

class Mammal(Animal):
    def breathe(self):
        print("Mammal is breathing.")

class Dog(Mammal):
    def bark(self):
        print("Dog is barking.")

my_dog = Dog()
my_dog.eat()  # Output: Animal is eating.
my_dog.breathe()  # Output: Mammal is breathing.
my_dog.bark()  # Output: Dog is barking.
```
